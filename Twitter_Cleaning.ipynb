```python
# -*- coding: utf-8 -*-
import os, re, json 
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
tweets_data = []

for line in open('SouthKorea.txt', "r"):  
    tweets_data.append(json.loads(line))

tweets = pd.io.json.json_normalize(tweets_data)  # normalize a dict/list of dicts into a flat table
```

```python
tweets.columns.tolist()
```

```python
headers = ['created_at','id', 'in_reply_to_status_id', 'place', 'coordinates', 'text','quoted_status.entities.hashtags', 'favorite_count', 'retweet_count', 'user.time_zone','lang']
tweets_df = tweets[headers]
tweets_df
```

# Save as CSV

```python
df = pd.DataFrame(tweets_df)
df.to_csv('/Users/DonginKim/Documents/Python/Social Media/SKorea.csv')
```
